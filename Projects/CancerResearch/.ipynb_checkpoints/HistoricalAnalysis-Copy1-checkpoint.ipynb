# The Evolution of Cancer Research

Cancer is currently one of the most well-funded research areas in the world, involving researchers from a wide variety of disciplinary backgrounds - but that hasn't always been the case. I have collected a dataset from the National Center for Biotechnology Information (NCBI) using their PubMed database. **The PubMed database is perhaps the largest and most widely used collection of biological research.** Our dataset includes every scientific research article in their database that includes the word "Cancer" in either the title or the abstract of the article as of Feb 12, 2020, which is **about 1.6 million scientific journal articles.** If you're interested, you can generate a similar dataset by [clicking here](https://www.ncbi.nlm.nih.gov/pubmed/?term=Cancer%5BTitle%2FAbstract%5D+NOT+%22QDF%22). 

The dataset includes the title and authors of each article, as well as some data about the publication date, which journal published the article, and some unique identifiers for the paper. Using that information, I'd like to better understand how cancer research has changed over time, and what it looks like today. **This is not meant to be a comprehensive study, but I believe that it does provide at least a glimpse into how cancer research as a field has developed over the last 200 years.**

Let's also check out how many journal articles we have in our dataset, and how they are spread out over time. Different entries seem to have recorded the year of publication in 3 different ways, so we need to make sure we grab the right information for each of the three formats, and they are sometimes in different columns of the dataset. I've arbitrarily numbered the different formats in my comments below.

```python
years = []
# Look at data from each article
for i in range(len(data)):
    if i > 0:
        article = data[i]
        short_details = article[4]
        pub_year = short_details.split(" ")[-1]
        # This is true if the date uses "format 1"
        if pub_year.isdigit():
            years.append(int(pub_year))
        # If the date was formatted differently, adjust the method for collecting it
        else:
            details = article[3]
            pub_year = details.split(";")[0][-4:]
            numbers = sum(character.isdigit() for character in pub_year)
            # True if date was in "format 2"
            if numbers == 4:
                years.append(int(pub_year))
            # True if date was in "format 3"
            else:
                pub_year = details.split(";")[0][-11:-7]
                years.append(int(pub_year))
        # Append extracted publication year back into article info, for later convenience
        article.append(pub_year)
        data[i] = article
            
# The number of years collected should be the number of lines in the dataset,
# minus the header line. Double check that we got them all.
if len(years) == len(data) - 1:
    print("All years collected - " + str(len(years)))

```

To be more quantitative, let's grab the title text for every decade since the 1940's (every decade with over 1,000 publications) for further analysis. I've printed out a few of the titles from the 1940's below.

```python
# Decades of interest - those with over 1,000 cancer publications
decades = [1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
data_subset = []
# Iterate over the selected decades
for i in range(len(decades)):
    data_subset.append([])
    # If the index isn't the final position in the decades list..
    if i < len(decades) - 1:
        # Collect article titles for all articles occuring in the specified decade
        # i indicates the decade index, article[-1] has publication year
        data_subset[i] = [article[0] for article in data[1:] if int(article[-1]) >= decades[i] and int(article[-1]) < decades[i+1]]
    # If the index is the final position in the decades list
    else:
        # Collect article titles for all articles occuring in the specified decade
        # Cut it off in 2019, so each subset covers the same number of years
        data_subset[i] = [article[0] for article in data[1:] if int(article[-1]) >= decades[i] and int(article[-1]) < 2020]
# data_subset is now a list, where each item in the list is a list of article titles from a specific decade
# Print the first few titles from the 1940's, just to double-check what is in data_subset
print(data_subset[0][0:6])
```

With our titles collected, it looks like we actually have some titles that are not in English. Although the cancer research in non-English languages does represent an important part of cancer research history and progress (and may even reveal if non-English researchers study different aspects of cancer), there is a high likelihood that keeping the non-English languages could reduce our confidence in our results by introducing duplicate information that we cannot meaningfully cross-reference (for example, 2 titles might be about identical topics, but being unable to translate the titles means they would get counted separately). We'll need to remove them from our dataset, at least for this part of the analysis. Removing non-English words will also likely remove specific words from the titles, such as specific gene or protein names, but if we have a specific gene of interest, we could go back and search for it in the original titles. 

Let's start that process by getting a set of english words, and adding in a few additional words of interest to that set.

```python
words = set(nltk.corpus.words.words()) # Set of English words, used to remove non-English words
# Add a couple common acronyms/abbreviations to the set of english words, so they don't
# get removed. This list is by no means exhaustive, and it is guaranteed that there will
# be acronyms or terms removed from the english titles - however, those terms are unlikely
# to be frequent enough to impact our results. The selection of words I add here could be
# a source of bias.
abbreviations = ["dna", "rna", "tcga", "seq",
                "acc", "brca", "crc", "blca",
                "cesc", "chol", "dlbc", "esca",
                "gbm", "hnsc", "kich", "kirc",
                "kirp", "laml", "lgg", "lihc",
                "luad", "lusc", "meso", "ucs",
                "uvm", "thym", "paad", "sarc",
                "pcpg", "ov", "prad", "skcm",
                "stad", "thca", "ucec", "ent"]
words.update(abbreviations)
# Also add a few'british' spellings of words that are likely to show up.
# Again, this won't be anywhere near exhaustive, but just contains a few 
# words that have a high likelihood of showing up in an article about cancer.
british = ["tumour", "tumours", "leukaemia", "oesophagus", "analyse", "ageing"]
words.update(british)
```

Now that we have our set of acceptable words together, let's go ahead and **remove the non-English words** from our article titles, and also **remove common 'stopwords'** like "and" or "the". While we're at it, let's also **lemmatize our words**. "Lemmatization" is essentially a way to normalize our words, for example, changing plurals to singular form, or changing the part of speech of a word. Doing this helps improve our ability to find similarities between titles (although due to the grammatical similarities among article titles, I suspect that lemmatization will not introduce many changes). Together, these **processing steps will help us condense our journal article titles down to their key ideas.**

There's quite a bit of code here, so **I've printed out a before and after version of the first couple of titles**, so you can see what they look like before and after these processing steps. The tools here aren't perfect, but it is enough processing to filter out a lot of artifacts from the data so that we can start asking meaningful questions.

```python
# function to convert nltk tag to wordnet tag
def nltk_tag_to_wordnet_tag(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:          
        return None

# Create a list to store the processed titles
decade_strings = []
# To help with printing the first couple titles
counter = 1
# For each decade in our subset of titles
for decade in data_subset:
    # Add a list to decade_strings, to hold data from a specific decade
    decade_strings.append([])
    # For each journal article title in this decade
    for title in decade:
        original_title = title
        # Convert to lowercase
        title = title.lower()
        # Remove punctuation - Note that journal article titles never use contractions
        title = title.translate(str.maketrans('', '', string.punctuation))
        # Tokenize the title and get the part of speech of each word
        nltk_tagged = nltk.pos_tag(nltk.word_tokenize(title))
        # Tuple of (token, wordnet_tag) for lemmatizing words
        wordnet_tagged = map(lambda x: (x[0], nltk_tag_to_wordnet_tag(x[1])), nltk_tagged)
        lemmatized_sentence = []
        # For each word and associated part of speech tag in the title
        for word, tag in wordnet_tagged:
            # If there is no available tag, append the token as is
            if tag is None:
                lemmatized_sentence.append(word)
            # If a part of speech tag was obtained, use it to lemmatize the word
            else:        
                lemmatized_sentence.append(lemmatizer.lemmatize(word, tag))
        # Remove stopwords
        # Multiple languages are included, but the stopwords in spanish have many similarities
        # to stopwords in other languages (de, la, el). It won't get everything, but it will help.
        # Also, remove tokens that are just numeric values
        filtered_sentence = [w for w in lemmatized_sentence \
                             if (not w in stopwords.words('english') \
                             and not w in stopwords.words('spanish')) and (not w.isdigit())]
        # Convert the filtered_sentence from a list into a string
        filtered_sentence = " ".join(filtered_sentence)
        # Remove non-English words that remained after the stopword filter.
        # Some words in non-English titles will still show up in the English word dictionary
        new_title = " ".join(w for w in nltk.wordpunct_tokenize(filtered_sentence) if w.lower() in words or not w.isalpha())
        # Append title to list of normalized titles from this decade
        curr_list = decade_strings[-1]
        curr_list.append(new_title)
        decade_strings[-1] = curr_list 
        # Print the before and after title, as an example of what
        # the processing steps did to our titles, for the first 6 titles
        if counter < 7:
            print("Original title " + str(counter) + ": " + original_title)
            print("Keywords from title " + str(counter) + ": " + new_title)
        counter += 1
```

We can see from the above original titles and associated key words, that the **natural language processings tools applied here have helped to simplify our journal article titles down to key words and ideas**. Just to point out some specific examples: In title 1, we see the stopword "of" was removed, we see punctuation and stopwords remove from title 2, in title 6 we see that we were able to remove most of the words from the non-English title, and in title 3 we see that the British word "tumours" was lemmatized to "tumour". We could translate the British words into the American spelling, but let's leave them for now.

Now that our titles have been largely processed, we can use those key words to get an idea about what **the biggest research topics from each decade** were. That information should help us better understand how cancer research has changed over time. As stated previously, our processing methods here are imperfect - we'll have to keep that in mind as we try to draw conclusions from our data. **Let's begin by looking at the top 10 most frequent words, by decade.**

```python
word_freqs = []
# Iterate over each decade
for i in range(len(decade_strings)):
    # Initialize a string for this decade
    full_string = ""
    # Concatenate all titles from this decade into a single string
    for title in decade_strings[i]:
        full_string += " " + title
    # Split the concatenated titles into individual words
    tokens = [t for t in full_string.split()]
    # Calculate the frequency of each word
    freq = nltk.FreqDist(tokens)
    # Save the frequencies
    word_freqs.append(freq)
    # Get the 10 most commond words
    top = freq.most_common(10)
    # Prepare a string containing the decade and the top 10 words from that decade
    out_string = str(decades[i]) + ": "
    for hit in top:
        out_string += hit[0] + " "
    # Print the top words from the decade
    print(out_string)
```

Obviously, we expected "cancer" to be the most common word in a dataset created by using the search word "cancer". Interestingly, **the word "breast" is the only other word that made the top 10 in every decade**, indicating the breast cancer has historically been a central part of cancer research. Another possible key take away is that **even though "cell" didn't make the top 10 until the 1980's, it is consistently in the top 5 once it appears, and consistently moves up the list** - this is likely due to technological advancements that allow researchers to study cancer at the cellular level.

To better visualize these trends, let's plot a few of these words, normalizing them by dividing the number of times they appear by the number of journal articles each decade. Let's focus on words that seem to be present only in either the early or later decades, to see what trends we can see.

```python
# Some important topics that appear to have notable trends based on the output above
words_of_interest = ["cell", "breast", "lung", "cervix", "treatment", "expression"]
freqs = []
# For each word of interest, calculate the number of times it appears in a decade
# and divide it by the number of publications in that decade. Multiply by 100.
for i in range(len(words_of_interest)):
    freqs.append([])
    for j in range(len(word_freqs)):
        value = 100 * (word_freqs[j][words_of_interest[i]] / len(decade_strings[j]))
        freqs[-1].append(value)

# Convert the above data into a dataframe for plotting
df=pd.DataFrame({'x': decades, 
                 "cell": freqs[0],
                 "breast": freqs[1],
                 "lung": freqs[2],
                 "cervix": freqs[3],
                 "treatment": freqs[4],
                 "expression": freqs[5]})
# RGB values for coloring our plot
colors = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),    
             (44, 160, 44), (152, 223, 138),(214, 39, 40), (255, 152, 150),    
             (148, 103, 189), (197, 176, 213)]
# Scale the RGB values to the [0, 1] range, which is the format matplotlib accepts.    
for i in range(len(colors)):    
    r, g, b = colors[i]    
    colors[i] = (r / 255., g / 255., b / 255.)  
# Plot the data
plt.plot( 'x', "cell", data=df, color = colors[0], linewidth=4)
plt.plot( 'x', "breast", data=df, color=colors[1], linewidth=4)
plt.plot( 'x', "lung", data=df, color=colors[2], linewidth=4)
plt.plot( 'x', "cervix", data=df, color=colors[3], linewidth=4)
plt.plot( 'x', "treatment", data=df, color=colors[4], linewidth=4)
plt.plot( 'x', "expression", data=df, color=colors[5], linewidth=4)
plt.ylabel("% of Publications")
plt.xlabel("Decade")
plt.title("Cancer Research Topics Over Time")
plt.legend();
```

These trends are pretty striking. **In 1940, only ~2% of cancer articles included the word "cell" in the title, but it now appears in over 20% of all cancer articles!** This, and the jump in "expression"-related articles indicate how technological advancements have impacted cancer research. Interestingly, the decade that saw the biggest increase (percentage-wise) in breast cancer research, also saw the first decreases for "treatment", "lung", and "cervix"-related papers, **perhaps suggesting that the 1970's represented a major shift in funding for cancer research** (moving towards breast cancer and away from lung and cervical cancer).  

Let's dig a little deeper into this dataset. Let's find the most common '3-word phrase' (trigram/3-gram) in each decade, and calculate how often we see that phrase across the decades. This will likely be very similar to the above information, but looking for 3-word phrases may help give some context to our findings. Our processing steps will be even more important here than they were in the previous section. Below, I've printed out the most common tri-gram for each decade.

```python
def get_top_n_trigram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(3, 3)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]

top_phrases = []
phrase_freqs = []
# Get the top 200 3-grams for each decade
for decade in decade_strings:
    freqs = get_top_n_trigram(decade, 200)
    phrase_freqs.append(freqs)

# List of lists, where each inner list is the counts of a given phrase in each decade.
# The number of inner lists will be the number of decades checked.
all_phrase_counts = []
decade_top_phrases = []
# Look at the top 200 3-grams for each decade
for i in range(len(phrase_freqs)):
    # Get the top 3-gram from the current decade
    top_decade_phrase = phrase_freqs[i][0][0]
    #print(top_decade_phrase)
    decade_counts = []
    # Count the frequency of that 3-gram in each of the decades,
    # if it wasn't in the top 200, it won't be found
    for j in range(len(phrase_freqs)):
        curr = phrase_freqs[j]
        value = 0
        for tup in curr:
            if tup[0] == top_decade_phrase:
                value = tup[1]
        # Normalize the value by dividing by the number of articles that decade
        value = 100 * (value / len(decade_strings[j]))
        decade_counts.append(value)
    # Save the counts from each decade, and the phrase used
    all_phrase_counts.append(decade_counts)
    decade_top_phrases.append(top_decade_phrase)
    
for i in range(len(decade_top_phrases)):
    print(str(decades[i]) + " most common tri-gram : " + decade_top_phrases[i]) 
```

And, let's go ahead and plot it like before. 

```python
# Convert the above data into a dataframe for plotting
df=pd.DataFrame({'x': decades, 
                 decade_top_phrases[0]: all_phrase_counts[0],
                 decade_top_phrases[1]: all_phrase_counts[1],
                 decade_top_phrases[2]: all_phrase_counts[2],
                 decade_top_phrases[3]: all_phrase_counts[3],
                 decade_top_phrases[4]: all_phrase_counts[4],
                 decade_top_phrases[5]: all_phrase_counts[5],
                 decade_top_phrases[6]: all_phrase_counts[6],
                 decade_top_phrases[7]: all_phrase_counts[7]})

# Plot the data. Commented out a few lines, since some decades had the same tri-gram
plt.plot( 'x', decade_top_phrases[0], data=df, color = colors[0], linewidth=4)
plt.plot( 'x', decade_top_phrases[1], data=df, color=colors[1], linewidth=4)
#plt.plot( 'x', decade_top_phrases[2], data=df, color=colors[2], linewidth=4)
plt.plot( 'x', decade_top_phrases[3], data=df, color=colors[3], linewidth=4)
plt.plot( 'x', decade_top_phrases[4], data=df, color=colors[4], linewidth=4)
plt.plot( 'x', decade_top_phrases[5], data=df, color=colors[5], linewidth=4)
#plt.plot( 'x', decade_top_phrases[6], data=df, color=colors[6], linewidth=4)
#plt.plot( 'x', decade_top_phrases[7], data=df, color=colors[7], linewidth=4)
plt.ylabel("% of Publications")
plt.xlabel("Decade")
plt.title("Cancer Research Topics Over Time")
plt.legend();
```

This plot is quite similar to the previous plot, but reveals a couple things. For one, we see that even though the word "lung" has shown up fairly consistently over the years, there has been a drastic increase in the tri-gram "cell lung cancer", likely indicating that specific lung cancer research areas have been on the rise (though this phrase accounts for only 2% of titles, while we saw the word "cell" 10 times as often). **It's also interesting, though perhaps counter-intuitive, that "treatment breast cancer" peaked decades before "human breast cancer", perhaps suggesting a transition in how breast cancer research is being approached.** Also, the previous plot showed that cervial cancer used to be more highly studied, and this plot suggests that uterine cancer is closely linked - though uterine cancer is often now referred to as endometrial cancer, so the nomenclature change may have also played a role.

It's possible that some of the earlier articles had more authors, but that they weren't listed in our dataset (instead listed as "et al.", for example). This may influence our conclusions slightly, though making it more difficult to find author information leads to the same problems as not listing authors, so it would represent similar problems.

To further investigate the collaborative nature of research, we can attempt to contruct a **social network using the lists of authors**. Since recent decades have tens of thousands of journal articles, and each one has several authors, I've chosen to put some restrictions on the authors that we'll include in our social networks, that will hopefully serve as a representative sample of trends in the field.

**We will focus our social network analysis on the articles from the 1970's and the 2000's, and will only include highly productive research that were listed as authors at least 10 times in a decade (averaging 1 publication per year) on papers about breast cancer.** The logic here is that 1) we've shown breast cancer research accounts for a similar percentage of work in those two decades, 2) limiting to highly productive authors will exclude many students from our analysis, while keeping most well-funded professionals, and 3) authors that have numerous, repeated publications on the same topic are more likely to have worked with other authors that have been studying that topic. This is obviously an incredibly small subset of our data, but it should be big enough to at least give us some information regarding the question we want to answer: How has collaboration and author recognition changed over time?

There's a lot of code here, but it's mostly repetitive, applying the different filters we've described. Ironically, the filtering requires more code than the actual graph creation.

```python
# SOCIAL NETWORK ANALYSIS ATTEMPT
# Large function..
# Input: Decade of interest, and our original data
def createGraph(decade, data):
    # Initialize some variables for keeping track of information
    node_dict = {} # Dictionary of nodes (authors), and how many publications they had in the decade
    nodes = set() # Set of nodes (authors), that meet out 10 publication threshold
    edges = [] # List of edges, indicating if 2 authors in out set of nodes have ever co-authored a paper
    G = nx.Graph() # Our social network graph
    articles = [] # A list for holding articles from the given decade
    
    # Count the number of times each author appears
    for article in data[1:]:
        year = int(article[-1]) # Year of publication
        if year - (year%10) == decade and "breast" in article[0].lower(): # If in our decade, and about breast cancer
            articles.append(article) # Add the article to the list of articles we're interested in
            author_list = article[2].split(",") # Split the string of authors into a list of authors
            if author_list[-1] == ".": # The final author often is followed by a period. Remove it
                author_list[-1] = author_list[-1][0:-1]
            for author in author_list: # Update our count on the number of times the author appeared in papers of interest
                if author in node_dict.keys():
                    node_dict[author.strip()] += 1
                else:
                    node_dict[author.strip()] = 1   
                    
    # Iterate through articles of interest. Similar to above
    # If an author had 10+articles, make them a node.
    # Only create edges between authors with 10+ articles
    for article in articles:
        year = int(article[-1]) # Year of publication
        author_list = article[2].split(",") # Get list of authors
        if author_list[-1] == ".": # Remove period from last author
            author_list[-1] = author_list[-1][0:-1]
        for author in author_list: # Iterate through author list
            author.strip() # Remove whitespace from te authors name
            # If the author had 10+ publications on breast cancer, and the author isn't actually just "et al."
            if author in node_dict.keys() and node_dict[author] >= 10 and (author != " et al." and author != " et al") :
                # Add the author as a node in our graph
                if author not in G:
                    G.add_node(author)
          
                    
    # Iterate through articles again..
    # Only create edges that made it through the filtering and were added to G
    for article in articles:
        year = int(article[-1]) # Publication year
        author_list = article[2].split(",") # Get author list
        if author_list[-1] == ".": #Remove period from last author
            author_list[-1] = author_list[-1][0:-1]
        # Create edges between authors from the article, if both of them 
        # met the filtering criteria to be added to our graph
        for i in range(len(author_list)): 
            for j in range(len(author_list)):
                if i < j:
                    if author_list[i] in G and author_list[j] in G:
                        edges.append((author_list[i], author_list[j]))
                                     
    # Remove duplicate edges
    edges_no_dup = list(set([ tuple(sorted(t)) for t in edges ]))
    edges = [list(elem) for elem in edges_no_dup]
    # Add edges to graph
    G.add_edges_from(edges)
    # Save our graph to a file for visualization
    #nx.write_gexf(G, str(decade) + "_network.gexf")
    return G

#Decades of interest
decades_for_graphs = [1970, 2000]
# Create social networks for our decades
# and print out basic information about the network
for decade in decades_for_graphs:
    G = createGraph(decade, data)
    print("Decade : " + str(decade)) # Decade of interest
    print("Number of authors : " + str(len(G.nodes)))
    print("Number of edges between authors : " + str(len(G.edges)))
    components = nx.connected_components(G)
    largest_component = max(components, key=len)
    subgraph = G.subgraph(largest_component)
    print("Network density (0-1) : " + str(nx.density(G)))
    print("Diameter of largest subnetwork : " + str(nx.diameter(subgraph)))
    print("-------------")

    
```

The above quantitative information is quite revealing. Obviously, there were more publications and more authors on those publications in the 2000's. **In the 1970's with 42 authors and only 32 edges, that means that many of the most prolific breast cancer researchers in the 1970's weren't working together. By contrast, the data from the 2000's indicates that most highly-productive breast cancer researchers worked with nearly 2 dozen other researchers that also fell in that category.** To better wrap our heads around this change, I've used some  network visualization software to visualize the networks generated here.
