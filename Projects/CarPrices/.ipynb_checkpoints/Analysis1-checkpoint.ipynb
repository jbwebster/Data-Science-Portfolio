```python
# Look at correlations between quantitative variables
dat_subset = data[['Year','Dealer_Price', 'City_MPG', 'Highway_MPG', 'Liters', 'V', 'Mileage', 'Customer_Review']]
plt.figure(figsize=(10,5))
c= dat_subset.corr()
sns.heatmap(c,annot=True)
# Interesting that having a bigger engine is the thing most highly correlated with being more expensive,
# customer reviews don't correlate with anything. Older cars have more miles on them. Bigger engines have
# terrible gas mileage. Basically confirms what we might expect.
# Interestingly, customer_reviews don't really correlate with anything - but their greatest positive
# correlation is with engine size, while their greatest negative correlations are with MPG. Odd

dat_subset.describe()
# Some possible outliers. 
# Consider that later (mainly in the Dealer_Price/MPG categories)
```

```python
# How to get counts of the categorical data
data_categorical = data[['NewUsed', 'Make', 'Model', 'Fuel_Type', \
                         'Exterior_Color', 'Interior_Color', 'Drivetrain', 'Transmission' ]]
print(data_categorical['NewUsed'].value_counts())
print('----')
print(data_categorical['Make'].value_counts())
print('----')
print(data_categorical['Fuel_Type'].value_counts())
```

```python
# How to create a bar graph of the categorical data counts
a = data_categorical['Make'].value_counts()
a.plot(kind='bar')
plt.xticks(rotation=90)
plt.show()
```

```python
# Example of how to do PCA.
# Here, I did PCA on the quantitative variables, then colored based on New vs Used
# Instead, consider changing New/Used to 0/1, then convert discretize Dealer_Price
# and see if the price groupings segregate at all. Consider making dummy variables
# for some of the other categorical variables

quant_features = ['Year','City_MPG', 'Highway_MPG', 'Liters', 'V', 'Mileage', 'Customer_Review']
# Extract the quantitative features, and standardize them
x = data[quant_features]
x = x.dropna() # OR IMPUTE INSTEAD. CURRENTLY DROPS ALL ROWS WITH NA VALUES
x = StandardScaler().fit_transform(x)

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
target = 'NewUsed'
y = data.loc[:,['NewUsed']].values
finalDf = pd.concat([principalDf, data[['NewUsed']]], axis = 1)

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['NEW', 'USED']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['NewUsed'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
# DROP THE MAX PRICE VEHICLE
data = data.drop(data['Dealer_Price'].idxmax())
```

```python
features = ['Drivetrain','Year','Dealer_Price', 'City_MPG', 'Highway_MPG', 'Liters', 'V', 'Mileage', 'Customer_Review']

dat_sub = data[features]
dat_sub = dat_sub.dropna()
# Note sure Drivetrain is the most interesting thing to have as a hue.
# Tempting to remove some features, just because of how big the plot is.
# This, in conjunction with the heatmap are both interesting

#SHOW WHOLE PLOT
#sns.pairplot(dat_sub, hue="Drivetrain");

# SHOW BOTTOM TRIANGLE (STACK OVERFLOW)
def hide_current_axis(*args, **kwds):
    plt.gca().set_visible(False)
    
# SHOW CORRELATION COEFFICIENT (STACK OVERFLOW)
def corrfunc(x,y, ax=None, **kws):
    """Plot the correlation coefficient in the top left hand corner of a plot."""
    r, _ = pearsonr(x, y)
    ax = ax or plt.gca()
    #ax = ax or plt.gca().set_visible(False)
    # Unicode for lowercase rho ()
    rho = '\u03C1'
    ax.annotate(f'{rho} = {r:.2f}', xy=(.5, .2), xycoords=ax.transAxes, size=16)

g = sns.pairplot(dat_sub) # Normal, color by drivetrain
# Can hide half the plot, or can add correlations
#g.map_upper(hide_current_axis) # Hide upper triangle
#g.map_lower(corrfunc); # Show correlation coefficients

g.map_upper(hide_current_axis)
#g.map_lower(corrfunc)
```
